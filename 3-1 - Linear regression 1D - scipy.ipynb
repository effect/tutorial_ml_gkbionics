## Linear regression with scipy

```python
import IPython
print(IPython.sys_info())
```

```python
import numpy as np
x = np.linspace(0, 10, 50)  # generate 50 values from 0 to 10
```

```python
y = 2.5 * x + 1.2
print(y)
```

```python
?np.linspace
```

```python
import matplotlib.pyplot as pl
%matplotlib inline
pl.plot(x, y, marker='s')
```

```python
import scipy as sc
noise = sc.randn(y.size)
print(noise)
```

```python
y_noise = y + noise
print(y_noise)
```

```python
pl.plot(x, y_noise, '.')
pl.xlabel('X')
pl.ylabel('Y')
```

```python
p = sc.polyfit(x, y_noise, 1)  # 1 is order (max degree of x)
print(p)
print(type(p))
```

```python
pl.plot(x, y_noise, '.')
pl.plot(x, p[0] * x + p[1], 'r-')
pl.xlabel('X')
pl.ylabel('Y')
```

```python
pl.plot(x, sc.polyval(p, x), 'r-')  # polyval instead of p[0] * x + p[1] by hand
```

```python
y_noise = y + noise * 10
pl.plot(x, y_noise, '.')
```

```python
p = sc.polyfit(x, y_noise, 1)
pl.plot(x, y_noise, '.')
pl.plot(x, sc.polyval(p, x), 'r-', label='Best fit')
pl.plot(x, y, 'k--', label='Input')
pl.legend(loc='best')
```

```python
pl.errorbar(x, y_noise, yerr=10, marker='.', linestyle='none')
pl.plot(x, sc.polyval(p, x), 'r-', label='Best fit')
pl.plot(x, y, 'k--', label='Input')
pl.legend(loc='best')
```
